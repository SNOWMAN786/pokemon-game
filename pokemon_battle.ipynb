```python
class Pokemon:
    def __init__(self, row):
        self.name        = row["name"]
        self.type1       = row["type1"]
        self.type2       = row["type2"]
        self.speed       = row["speed"]
        self.defense     = row["defense"]
        self.sp_defense  = row["sp_defense"]
        self.attack      = row["attack"]
        self.sp_attack   = row["sp_attack"]
        self.base_total  = row["base_total"]
        self.max_hp      = row["hp"]
        self.hp          = row["hp"]
        self._row        = row

    def get_type_multiplier(self, attack_type: str) -> float:
        col = TYPE_TO_COL.get(attack_type)
        if col and col in self._row.index:
            return float(self._row[col])
        return 1.0

    def choose_attack(self) -> str:
        """Player picks between type1 and type2 attack."""
        print(f"\n  [1] {self.type1}   [2] {self.type2}")
        while True:
            try:
                pick = int(input("  Choose attack (1 or 2): "))
                if pick == 1: return self.type1
                if pick == 2: return self.type2
                print("  Enter 1 or 2.")
            except ValueError:
                print("  Invalid — enter a number.")

    def ai_choose_attack(self, opponent: "Pokemon") -> str:
        """
        AI attack selection.
        Uses the trained ML model if available, otherwise falls back to random.
        The features here MUST match exactly what train_ai.py trained on.
        """
        if AI_MODEL is not None:
            mult1 = opponent.get_type_multiplier(self.type1)
            mult2 = opponent.get_type_multiplier(self.type2)

            features = np.array([[
                self.hp / self.max_hp,      # ai_hp_ratio
                self.attack,                # ai_attack
                self.sp_attack,             # ai_sp_attack
                self.speed,                 # ai_speed
                opponent.hp / opponent.max_hp,  # player_hp_ratio
                opponent.defense,           # player_defense
                opponent.sp_defense,        # player_sp_defense
                mult1,                      # mult_type1
                mult2,                      # mult_type2
            ]])

            prediction = AI_MODEL.predict(features)[0]  # 0 → type1, 1 → type2
            return self.type1 if prediction == 0 else self.type2

        # fallback if model not loaded
        return np.random.choice([self.type1, self.type2])

    def __repr__(self):
        return (f"{self.name} | {self.type1}/{self.type2} | "
                f"HP:{self.hp:.0f}/{self.max_hp:.0f} | "
                f"ATK:{self.attack} SPD:{self.speed}")
```

```python
def health_bar(poke: Pokemon):
    bar_length   = 20
    ratio        = max(poke.hp, 0) / poke.max_hp
    filled       = int(bar_length * ratio)
    bar          = "█" * filled + "-" * (bar_length - filled)
    print(f"  {poke.name:<20} HP: [{bar}] {max(poke.hp, 0):.0f}/{poke.max_hp:.0f}")
```

```python
# ── Battle loop ───────────────────────────────────────────────────────────────
print("\n══════════ BATTLE START ══════════\n")

while player_pokemon.hp > 0 and ai_pokemon.hp > 0:

    health_bar(player_pokemon)
    health_bar(ai_pokemon)

    # — Player turn ——————————————————————————————————————
    print(f"\n⚔  {player_pokemon.name}'s turn:")
    attack_type     = player_pokemon.choose_attack()
    multiplier      = ai_pokemon.get_type_multiplier(attack_type)
    attack          = Attack(attack_type)
    damage          = attack.damage(player_pokemon, ai_pokemon)  # ← removed multiplier
    ai_pokemon.hp  -= damage

    eff = {0.0: "has no effect", 0.5: "is not very effective", 2.0: "is super effective!"}.get(multiplier, "")
    print(f"  ➤ {player_pokemon.name} used {attack_type}! "
          f"{'(' + eff + ')  ' if eff else ''}→ {damage:.1f} dmg  (×{multiplier})")

    if ai_pokemon.hp <= 0:
        print(f"\n  💀 {ai_pokemon.name} fainted!")
        break

    # — AI turn ——————————————————————————————————————————
    print(f"\n🤖  {ai_pokemon.name}'s turn:")
    ai_type = ai_pokemon.ai_choose_attack(player_pokemon)
    ai_mult        = player_pokemon.get_type_multiplier(ai_type)
    attack         = Attack(ai_type)
    damage         = attack.damage(ai_pokemon, player_pokemon)  # ← fixed attacker/defender order
    player_pokemon.hp -= damage

    eff = {0.0: "has no effect", 0.5: "is not very effective", 2.0: "is super effective!"}.get(ai_mult, "")
    print(f"  ➤ {ai_pokemon.name} used {ai_type}! "
          f"{'(' + eff + ')  ' if eff else ''}→ {damage:.1f} dmg  (×{ai_mult})")

    if player_pokemon.hp <= 0:
        print(f"\n  💀 {player_pokemon.name} fainted!")
        break

    print()

# ── Result ───────────────────────────────────────────────────────────────────
print("\n══════════ RESULT ══════════")
if player_pokemon.hp > 0:
    print(f"  🏆 You win! {player_pokemon.name} survived with {player_pokemon.hp:.0f} HP.")
else:
    print(f"  You lost. {ai_pokemon.name} wins!")
```
